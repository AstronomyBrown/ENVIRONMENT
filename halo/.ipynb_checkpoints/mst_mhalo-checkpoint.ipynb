```python
%matplotlib inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt

path = '/Users/thbrown/AA_project/TOBY_STACKING/SAMPLES/FULL/masters/'

# read in sample table

hdulist = pyfits.open(path+'sample_grp.fits')
tbdata = hdulist[1].data
ID = tbdata['ID']
ms = tbdata['lgMst_median']
mh = tbdata['logMh_Mst']
flag = tbdata['flag_mstar'] #1 = BGG, 2 = satelite
grp_id = tbdata['grp_id']
ngal = tbdata['Ngal']
```

```python
# write new column containing bgg M* only. Satellites are assigned their accompanying bgg mass.
print ms
bgg_ms = tbdata['lgMst_median']
for i in range(len(ID)):
    if flag[i]==2:
        if len(ms[(grp_id==grp_id[i]) & (flag ==1)])==1:
            bgg_ms[i] = ms[(grp_id==grp_id[i]) & (flag ==1)][0]
        else:
            bgg_ms[i] = -99
```

```python
bin_size = 0.1; min_edge = 8; max_edge = 12
N = (max_edge-min_edge)/bin_size; Nplus1 = N + 1
bin_list = np.linspace(min_edge, max_edge, Nplus1)

plt.hist(bgg_ms, bins=bin_list)
plt.ylabel('N')
plt.show()

plt.hist(ms, bins=bin_list)
plt.ylabel('N')
plt.xlabel('log stellar mass')
plt.show()
```

```python
# assign and histogram M*/Mhalo. Set values for satellites without central in sample to -99

ms_mh = 10**bgg_ms / 10**mh

ms_mh[(ms_mh==0.)] = -99

plt.hist(ms_mh, bins=10 ** np.linspace(np.log10(0.0001), np.log10(.1), 50))
plt.ylabel('N')
plt.xlabel('M*/Mhalo')
plt.gca().set_xscale("log")
plt.show()
```

```python
# write to file.

# set col formats
fmt = "%i %2.5f"

path = '/Users/thbrown/AA_project/TOBY_STACKING/SAMPLES/FULL/constituent_tables/'
wfile = 'ID_ms_mh.dat'

txttab = np.column_stack([ID, ms_mh])
np.savetxt(path+wfile, txttab, delimiter="    ", fmt=fmt)

print 'file written to ' + path + wfile
```

```python
# SMF (see Behroozi+10,+13)
plt.semilogy(mh, ms_mh, '.', markersize = 4)
plt.title('SMF vs. Mhalo')
plt.ylabel('M*/Mhalo')
plt.xlabel('log Mhalo')
plt.ylim(0.0001,0.1)
plt.xlim(10,15)
```

```python
# Histograms in central, satellite and isolated bins
# plt.subplots_adjust(hspace=0.4)

# plt.subplot(221)
plt.hist(ms_mh, bins=10 ** np.linspace(np.log10(0.0001), np.log10(.1), 50))
plt.title('all galaxies')
plt.ylabel('N')
plt.xlabel('M*/Mhalo')
plt.gca().set_xscale("log")
plt.show()

plt.hist(ms_mh[(flag==2)], bins=10 ** np.linspace(np.log10(0.0001), np.log10(.1), 50))
plt.title('centrals')
plt.ylabel('N')
plt.xlabel('M*/Mhalo')
plt.gca().set_xscale("log")
plt.show()

plt.hist(ms_mh[(flag==1) & (ngal > 1)], bins=10 ** np.linspace(np.log10(0.0001), np.log10(.1), 50))
plt.title('satellites')
plt.ylabel('N')
plt.xlabel('M*/Mhalo')
plt.gca().set_xscale("log")
plt.show()

plt.hist(ms_mh[(ngal==1)], bins=10 ** np.linspace(np.log10(0.0001), np.log10(.1), 50))
plt.title('isolated')
plt.ylabel('N')
plt.xlabel('M*/Mhalo')
plt.gca().set_xscale("log")
plt.show()
```
